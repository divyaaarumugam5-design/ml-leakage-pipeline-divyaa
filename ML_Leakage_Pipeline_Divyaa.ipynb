Task 1 — Reproduce and Identify Leakage Using the synthetic dataset generated below, scale the features on the entire dataset before splitting, train a Logistic Regression model, and report train and test accuracy. Identify what is wrong with this approach.

```python
#Task 1 — Reproduce and Identify Leakage
from sklearn.datasets import make_classification
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
X, y = make_classification(n_samples=1000, n_features=10, random_state=42)
scale=StandardScaler()
X_scale=scale.fit_transform(X)
X_train,X_test,y_train,y_test=train_test_split(X_scale,y,test_size=0.25,random_state=42)

model=LogisticRegression()
model.fit(X_train,y_train)
print("Train Accuracy:", model.score(X_train,y_train))
print("Test Accuracy:",model.score(X_test,y_test))


```

From the above task 1 we can see Training accuracy and testing accuracy is approximately same which implies **Data Leakage ** . Scaling (transformation) occurs befoe we splitting dataset which results in the leakage of test dataset.

Task 2 — Fix the Workflow Using a Pipeline Refactor the code from Task 1 using a Pipeline that combines StandardScaler and LogisticRegression. Split the data first using train_test_split, then run 5-fold cross-validation using cross_val_score. Report mean accuracy and standard deviation.

```python
#Task 2 — Fix the Workflow Using a Pipeline
from sklearn.datasets import make_classification
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score
X, y = make_classification(n_samples=1000, n_features=10, random_state=42)
#splitting
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
#Pipeline
pipe=Pipeline([
      ('scale',StandardScaler()),
      ('model',LogisticRegression())
      ])
#Cross Validation
scores=cross_val_score(pipe,X_train,y_train,cv=5,scoring="accuracy")
# Mean accuarcy and Standard Deviation
print("Mean accuracy:",round(scores.mean(),2))
print("Std:",round(scores.std(),2))
```

Task 3 — Experiment with Decision Tree Depth Train a DecisionTreeClassifier at max_depth values of 1, 5, and 20 using the same train-test split from Task 2. Record train and test accuracy for each depth in a table and briefly explain which depth best balances fit and generalization.

```python
#Task 3 — Experiment with Decision Tree Depth
from sklearn.datasets import make_classification
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
X, y = make_classification(n_samples=1000, n_features=10, random_state=42)

#splitting
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)

for depth in [1,5,20]:
  model=DecisionTreeClassifier(max_depth=depth,random_state=42)
  model.fit(X_train,y_train)
  train_acc=model.score(X_train,y_train)
  test_acc=model.score(X_test,y_test)
  print(f"Depth:{depth}|Train accuracy:{round(train_acc,2)}|Test accuracy:{round(test_acc,2)}")
```

Depth 1 with training accuracy and testing accuracy shows underfitting. Depth 5 from training and testing accuracy we can analyse this as correct fitting and with depth 20 while training it give 100% and while testing dataset shows low accuracy which can be overfitting
